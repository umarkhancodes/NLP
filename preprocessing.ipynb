```python
train_v=train_nn[['pros','cons','rating_diversity']]
train_v['reviews']=train_v['pros']+'. '+train_v['cons']
train_v=train_v[['reviews','rating_diversity']]
train_v
```

```python
import nltk
import string
import fasttext
import contractions 

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords, wordnet 
from nltk.stem import WordNetLemmatizer

import matplotlib.pyplot as plt

plt.xticks(rotation=70)
pd.options.mode.chained_assignment=None
pd.set_option('display.max_colwidth',100)
%matplotlib inline
```

```python
train_v['no_contract']=train_v['reviews'].apply(lambda x : [contractions.fix(word) for word in x.split()])
train_v
```

```python
train_v['rating_description_str']=[' '.join(map(str,l)) for l in train_v['no_contract']]
train_v
```

```python
train_v['tokenized']=train_v['rating_description_str'].apply(word_tokenize)
train_v
```

```python
# import nltk
# nltk.download('punkt')
```

```python
train_v['lower']=train_v['tokenized'].apply(lambda x: [word.lower() for word in x])
train_v
```

```python
punc=string.punctuation
train_v['no_punc']=train_v['lower'].apply(lambda x : [word for word in x if word not in punc])
train_v
```

```python
from autocorrect import Speller
```

```python
spell = Speller(lang='en')
train_v['spell']=train_v['no_punc'].apply(lambda x: [spell(word) for word in x])
train_v
```

```python
train_v['no_num']=train_v['spell'].apply(lambda x: [word for word in x if word.isalpha()])
train_v
```

```python
stop_words=set(stopwords.words('english'))
train_v['stopwords_removed']=train_v['no_num'].apply(lambda x : [word for word in x if word not in stop_words])
train_v
```

```python
# nltk.download('stopwords')
```

```python
train_v['pos_tags']=train_v['stopwords_removed'].apply(nltk.tag.pos_tag)
train_v
```

```python
# nltk.download('averaged_perceptron_tagger')
```

```python
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    if tag.startswith('V'):
        return wordnet.VERB
    if tag.startswith('N'):
        return wordnet.NOUN
    if tag.startswith('R'):
        return wordnet.ADV
    else :
        return wordnet.NOUN
train_v['wordnet_pos']=train_v['pos_tags'].apply(lambda x : [(word,get_wordnet_pos(pos_tag)) for (word,pos_tag) in x])
train_v
    
```

```python
# nltk.download('wordnet')
```

```python
wnl=WordNetLemmatizer()
train_v['lemmatized']=train_v['wordnet_pos'].apply(lambda x : [wnl.lemmatize(word,tag) for word,tag in x])
train_v
```

```python
def short_words(words):
    return [word for word in words if len(word)>2]
train_v['length']=train_v['lemmatized'].apply(short_words)
train_v
```

```python
train_v['final_review']=train_v['length']
```

```python
train_final=train_v[['final_review','rating_diversity']]
```
